````python
import numpy as np
import matplotlib.pyplot as plt
````

二つの中心点（2.0, 54.50)と（4.3, 80.0）を```mus```に設定する<br>
それぞれの中心点からの分散を```covs```にて設定する。<br>
それぞれの出現確率を```phis```に設定する<br>
これら値は、ここではトップダウン的に「決められた」値として参照するが、本来はstep5でのEMアルゴリズム(テキストp.87以降)にて求められているとのこと。

````python
mus = np.array([[2.0, 54.50],
                [4.3, 80.0]])

covs = np.array([[[0.07, 0.44],
                  [0.44, 33.7]],
                 [[0.17, 0.94],
                  [0.94, 36.00 ]]])

phis = np.array([0.35, 0.65])
````

````python
def sample():
    k = np.random.choice(2, p=phis)
    mu, cov = mus[k], covs[k]
    x = np.random.multivariate_normal(mu, cov)
    return x
````

````python
N = 500
xs = np.zeros((N, 2))
for i in range(N):
     xs[i] = sample()
````

````python
plt.scatter(xs[:,0], xs[:,1], color='orange', alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````
