```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-5,5,100)
y_relu = torch.relu(x)
y_sigmoid = torch.sigmoid(x)
y_tanh = torch.tanh(x)
y_elu = F.elu(x)
y_gelu = F.gelu(x)
y_rrelu = F.rrelu(x)
y_selu = F.selu(x)
```

```python
plt.subplots_adjust(hspace=0.6)
plt.subplot(3,2,1)
plt.title("sigmoid")
plt.plot(x,y_sigmoid)
plt.subplot(3,2,2)
plt.title("relu")
plt.plot(x,y_relu)
plt.subplot(3,2,3)
plt.title("elu")
plt.plot(x,y_elu)
plt.subplot(3,2,4)
plt.title("gelu")
plt.plot(x,y_gelu)
plt.subplot(3,2,5)
plt.title("rrelu")
plt.plot(x,y_rrelu)
plt.subplot(3,2,6)
plt.title("selu")
plt.plot(x,y_selu)
plt.show()
```
