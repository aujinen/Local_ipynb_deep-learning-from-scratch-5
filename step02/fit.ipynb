【注意】オリジナルコードとして同梱されている「notebooks」フォルダ内のコードをローカルで直接実行すると同一フォルダ内に該当するデータファイルが存在しないのでエラーが発生する。<br>

````python
import os
import numpy as np
import matplotlib.pyplot as plt
````

````python
#=# path = os.path.join(os.path.dirname(__file__), 'height.txt')
path = os.path.join(os.getcwd(), 'height.txt')
xs = np.loadtxt(path)

mu = np.mean(xs)
sigma = np.std(xs)
````

````python
print (f'{mu=}')
print (f'{sigma=}')
````

上記のサンプルデータから得られる平均と標準偏差を元にして得られる正規分布の曲線（最尤推定にて演繹的に得られる曲線）の定義<br>
normal関数の引数内にある```mu=0```と```sigma=1```は、関数呼び出し時にパラメータが省略された場合にデフォルトとなる値。

````python
# normal distribution
def normal(x, mu=0, sigma=1):
    y = 1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-(x - mu)**2 / (2 * sigma**2))
    return y
````

````python
x = np.linspace(150, 190, 1000)
y = normal(x, mu, sigma)
````

````python
# plot
plt.hist(xs, bins='auto', density=True)
plt.plot(x, y)
plt.xlabel('Height(cm)')
plt.ylabel('Probability Density')
plt.show()
````

normal関数で、引数の```mu```のみ推定値を用い、```sigma```を省略して重ね合わせた場合

````python
x = np.linspace(150, 190, 1000)
y = normal(x,mu)
````

````python
# plot
plt.hist(xs, bins='auto', density=True)
plt.plot(x, y)
plt.xlabel('Height(cm)')
plt.ylabel('Probability Density')
plt.show()
````
