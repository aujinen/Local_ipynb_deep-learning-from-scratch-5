````python
import os
import numpy as np
import matplotlib.pyplot as plt
````

#### パラメータ推定
最尤推定にて正規分布のパラメータを得る

````python
#=# path = os.path.join(os.path.dirname(__file__), 'height.txt')
path = os.path.join(os.getcwd(), 'height.txt')
xs = np.loadtxt(path)
mu = np.mean(xs)
sigma = np.std(xs)

print (f'{mu=}')
print (f'{sigma=}')
````

### 【注意】
テキストでは、「パラメータ推定は『学習』とも呼ばれています」と記述していますが、通常では見かけない表現であり、この場所においては誤解を招く表現となるので、注意してください。<br>
いわゆる「機械学習」の『学習』や「深層学習」の『学習』は、目標（正答）となる出力との誤差を最小にするように、内部パラメータを修正しつつ、繰り返す行為（都度、よりよい方向へと『学習』していく行為）として意味づけられています。<br>
この場所では、統計学的な最尤推定として定式化された正規分布を用いて推定を行う段階であり、『学習』とは異なります。<br>
たとえ、ここで利用している「正規分布」という最尤推定方法が「機械学習」や「深層学習」で学習した結果得られたモデルだったとしても、今回のようにサンプルデータ等の実際のデータに対する推定に用いる段階では「学習後のモデルをもちいて」という表現にすべきところです。

#### テキストp.37最後のカラム相当分のコード例

````python

samples = np.random.normal(mu, sigma, 10000)

plt.hist(xs, bins='auto', density=True, alpha=0.7, label='original')
plt.hist(samples, bins='auto', density=True, alpha=0.7, label='generated')
plt.xlabel('Height(cm)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
````
